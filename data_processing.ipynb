```python
paths = ["./hurricane_sandy/" + dir for dir in os.listdir("hurricane_sandy")]
xr.open_mfdataset(paths)
```

```python
index = 0
for nc_file in paths:
    dataset = xr.open_dataset(nc_file)
    sla = dataset.variables["sla"][:]
    latitude = dataset.variables["latitude"][:]
    longitude = dataset.variables["longitude"][:]
    dataset.close()

    date = nc_file.split('_')[3]

    # Create a meshgrid for plotting
    lon, lat = np.meshgrid(longitude, latitude)
    
    # Plot the 'sla' variable
    plt.figure(figsize=(10, 6))
    cbarticks=np.arange(-1.25,1.25,0.25)
    plt.contourf(lon, lat, sla[0, :, :], cbarticks, cmap="bwr", vmin=-1.25,vmax=1.25)
    cbar = plt.colorbar(label="Sea Level Anomaly (m)")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.ylim()
    plt.title(f"Sea Level Anomaly on {date}")
    plt.savefig(f'./hurricane_sandy_sla/image{index}')
    
    index += 1
    

```

```python
from matplotlib.animation import FuncAnimation

nframes = 60
plt.subplots_adjust(top=1, bottom=0, left=0, right=1)

def animate(i):
    im = plt.imread('./hurricane_sandy_sla/image'+str(i)+'.png')
    plt.imshow(im)

anim = FuncAnimation(plt.gcf(), animate, frames=nframes, 
                     interval=(2000.0/nframes))
anim.save('output.gif', writer='imagemagick')
```
